```python
from analysis_tools import *
import math
import random
import matplotlib.pyplot as plt
import pickle as pkl
import numpy as np
```

Load Data

```python
with open('semantic_point_cloud.pickle', 'rb') as file:
    full_data = pkl.load(file)

groundings = full_data[0][1:]
T = np.transpose(full_data)
words = T[0]
print(words)
print(f"type {type(full_data)}, shape {full_data.shape}")
print(groundings)
```

Filter the data including select groundings: Socialness, OLD, PLD, Gustatory, Mouth

```python
columns_to_keep = [0,1,3,4,7,15]
columns_to_keep = np.array(list(dict.fromkeys(columns_to_keep)))

# Use advanced indexing to select the desired columns
S = full_data[:, columns_to_keep]
S = np.transpose(S)
```

Compute Jaccard curves for selected groundings. Compute the numerical Jaccard invariant with param = 1 (Total area under the curve)

```python
param = 1
n = len(S)
areas = np.zeros((n,n))
for i in range(1,n):
        G1 = S[i]
        #G is the columns i.e groundings
        L1 = normalise_sort([words[1:],G1[1:]])
        for j in range(i + 1, n):
            G2 = S[j]
            L2 = normalise_sort([words[1:],G2[1:]])
            if G1[0] != G2[0]:
                x,y = percent_compare_ordered(L1, G1[0], L2, G2[0])
                distance = area_estimation(x, y)
                areas[i, j] = distance
                areas[j, i] = distance
areas = np.delete(areas, 0, axis=0)
areas = np.delete(areas, 0, axis=1)
```

Display distances as heatmap

```python
plt.imshow(areas, cmap='hot', interpolation='nearest')
plt.colorbar()  # Add a colorbar to show the scale
plt.title("Inverse distances between groundigns")
plt.show()
```

Combine groundings Valance and Arousal.

```python
valence = T[17]
arousal = T[18]
interoceptive = T[9]
combined = combine_raw([words[1:],[linear_transform(value,min(arousal[1:]),max(arousal[1:]),0,1) for value in arousal[1:]]], [words[1:],[linear_transform(value,min(valence[1:]),max(valence[1:]),0,1) for value in valence[1:]]], "mean")
emotion = np.array(["Emotion"]+list(combined[1]))

G = [valence, arousal, interoceptive, emotion]
```

Compute Jaccard curves for new groundings.

```python

m = len(G)
areas = np.zeros((m,m))
for i in range(0,m):
    G1 = G[i]
    #G is the columns i.e groundings
    L1 = normalise_sort([words[1:],G1[1:]])
    for j in range(i+1,m):
        G2 = G[j]
        L2 = normalise_sort([words[1:],G2[1:]])
        if G1[0] != G2[0]:
            x,y = percent_compare_ordered(L1, G1[0], L2, G2[0])
            distance = area_estimation(x, y)
            areas[i, j] = distance
            areas[j, i] = distance
```

Heatmap

```python
plt.imshow(areas, cmap='hot', interpolation='nearest')
plt.colorbar()  # Add a colorbar to show the scale
plt.title("Inverse distances between groundigns")
plt.show()
```
